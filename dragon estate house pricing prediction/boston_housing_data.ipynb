```python
housing=strat_train_set.drop("MEDV",axis=1)
housing_labels=strat_train_set["MEDV"].copy()
```

## MISSING ATTRIBUTES

```python
# TO TAKE CARE OF MISSING ATTRIBUTES , YOU HAVE THREE OPTIONS
# 1...GET RID OF TH MISSING DATAA POINTS
# 2...GET RID OF THE WHOLE ATTRIBUTE
# 3...SET THE VALUE TO SOME VALUE(0,MEAN OR MEDIAN)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="median")
imputer.fit(housing)
```

```python
imputer.statistics_.shape
```

```python
# kisi b column m missing  value ki jgh median fix ho jaye
X=imputer.transform(housing)
housing_tr=pd.DataFrame(X,columns=housing.columns)
housing_tr.describe()
```

## SCIKIT-LEARN DESIGN

Threer types of objects 

1...Estimators-> It estimates some paramter based on a dataset. Eg imputer . It has a fit method and transform method . Fit method fits the dataset and calculates the internal parameters

2...Transformers->transform method takes input and returns output based on the learnings from the fit(). It also has a convience function calles fit_transform() whoch fits and then transform

3...Predictors-> eg-LinearRegression . fit() and predict() are two common fucntions . It also gives score () function which will evaluate the predictions

## CREATING A PIPELINE

 FEATURE SCALING --->> TWO TYPES :
 
1... min-max scaling/normalization ->>(val-min)/(max-min)
class calles MinMaxScaler for this
2...Standardization->>(value-mean)/std
class provides Standard Scaler for this 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([
    ('imputer' , SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler()),])
```

```python
#housing_num_tr--->numpy array not a dataset 
housing_num_tr=my_pipeline.fit_transform(housing)
```

## SELECTIING A DESRIED MODEL FOT THE DRAGON REAL ESTATES


```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model=LinearRegression()
model=RandomForestRegressor()
# model=DecisionTreeRegressor()
model.fit(housing_num_tr,housing_labels)
```

```python
some_data=housing.iloc[:5]
```

```python
some_labels=housing_labels.iloc[:5]
```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

## EVALUATING THE MODEL

```python
from sklearn.metrics import mean_squared_error
housing_predictions=model.predict(housing_num_tr)
lin_mse=mean_squared_error(housing_labels,housing_predictions)
lin_rmse=np.sqrt(lin_mse)
```

```python
lin_rmse   #discard the model as it gives high error and if some model gives error as 0 sometimes its bcz of overfitting of the data 
```

## using better evaluation technique-cross validation

```python
from sklearn.model_selection import cross_val_score 
scores=cross_val_score(model,housing_num_tr,housing_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores=np.sqrt(-scores)
```
